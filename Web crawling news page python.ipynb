제 전공은 컬처앤테크놀로지융합전공이고 콘텐츠를 공부하는 전공이기에 '화제성'은 굉장히 중요한 포인트 입니다.
저는 주로 화제성 높은 이슈들을 뉴스 기사들을 통해 확인하고 콘텐츠 제작에 반영합니다.

그래서 저는 조회수를 기준으로 언론사별 상위 5개 기사를 게시하는 네이버 랭킹 뉴스 페이지(https://news.naver.com/main/ranking/popularDay.naver)를 웹크롤링 해보았습니다.

일반적으로 뉴스 메인 페이지들은 다양한 반영 기준(분야, 최신성 등)을 사용해 라이브와 심층 기획 등의 섹션으로 나누어 뉴스를 보여준다면, 
네이버 랭킹 뉴스 페이지는 '조회수', 즉 화제성을 기준으로 언론사 별 상위 5개 기사를 실시간 업데이트 후 게시합니다. 
또 언론사 별로 나누어져 있기 때문에 각 언론사에서 화제성이 높은 기사들의 분야들을 비교해 볼 수 있으며 
이를 바탕으로 언론사 별 주목 받는 이슈들을 분석하기에 적절합니다.

다만 해당 사이트 페이지는 언론사의 종류와 기사 개수가 너무 방대합니다.
그리고 화제인 기사들을 한 눈에 볼 수 있지만 본인이 원하는 키워드가 있는 기사들을 따로 필터링 할 수는 없습니다.

따라서 저는 콘텐츠 제작시 반영할 수 있는 화제성 높은 언론사 10개와 각 언론사들의 상위 5개 기사들을 제가 원하는 키워드에 따라 확인 후 
별도로 출력 할 수 있는 기능을 추가하였습니다.

html 코드 구조를 보고 언론사 이름을 나타내는 태그와 기사 제목의 태그를 확인해 따로 분류했고,
출력과 동시에 딕셔너리에 언론사 별로 상위 기사 5개들을 매칭하려 했습니다.
그러나 파싱 된 내용들은 모두 html코드와 태그들을 포함하고 있어 텍스트만 따로 추출하기 위해 
리스트에 언론사와 기사 제목을 분리해 리스트 형식으로 저장했고
이후 해당 리스트를 활용해 딕셔너리에 정리해 옮겨 담았습니다.

해당 과제는 단순히 내용을 분류해 저장 하는 것을 넘어 
프로그램처럼 구현해 유저가 원하는 키워드를 작성하면 해당 키워드가 들어간 기사들을 언론사의 이름과 함께 보여주고
반복문을 사용해 자신이 원하는 횟수 만큼 여러 키워드를 입력해 보면서 관련 기사 제목들을 찾아볼 수 있게 구현하였습니다.
이를 통해 저는 각 언론사 별로 제가 검색하고자 하는 키워드가 어느 정도의 화제성을 지니는지, 또 제목을 통해 해당 키워드가 어떤 관점에서 다뤄지는지 
한 눈에 쉽게 분석할 수 있었습니다.

```python
!pip install bs4
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://news.naver.com/main/ranking/popularDay.naver'
#해당 사이트 페이지의 코드 구조를 확인하기 위해 코드 블럭을 분리 시켜 출력 결과를 따로 출력
content = requests.get(link)
print (content.text)
```

```python
website_code = content.text

parsing=BeautifulSoup(website_code,'html.parser')

media_name = parsing.find_all('strong',{"class":"rankingnews_name"}) #언론사의 이름들 분류
top_news = parsing.find_all('a',{"class":"list_title nclicks('RBP.rnknws')"}) #뉴스 기사 제목들 분류

#위 파싱된 내용들은 모두 html 코드를 포합하고 있어 텍스트만 따로 분류해줄 필요가 있음
media_list=[] #언론사의 이름만 따로 저장
medianews_list = [] #뉴스 기사만 따로 저장

#뉴스 제목을 저장한 리스트가 언론사 번호 구별 없이 저장 됨. 그래서 html을 통해 기사들을 확인해보면
#언론사 하나에 기사 5개씩 묶인다는 것을 확인 가능.
#실제 사이트 페이지 내용과 코드 출력 결과물을 비교 확인하여 언론사별 상위 5개 기사가 정확히 매칭 된다는 것을 확인함.
count = 0 
cut_count=5

for media in media_name[:10]: #언론사 개수가 너무 많아서 가장 먼저 검색되는 10개의 언론사만 추출
    print(media.get_text(strip=True)+" 상위 5개 기사") #.get_text(strip=True)를 통해 텍스트 이외의 코드나 공백 제거
    media_list.append(media.get_text(strip=True))
    
    for i,news in enumerate(top_news[count:cut_count],1): #언론사 별로 상위 5개 기사들을 순위 번호와 함께 5개만 추출
        print(str(i)+". "+news.get_text(strip=True))
        medianews_list.append(news.get_text(strip=True))
        
    print("\n")
    #다음 5개 기사들을 추출하기 위한 범위 번호를 5 단위로 업데이트
    count+=5
    cut_count+=5

#딕셔너리를 통해 각 언론사와 해당 언론사의 상위 기사 5개 매치
news_dict={}
num_count = 0
for name in media_list:
    news_dict[name]=medianews_list[num_count:num_count+5]
    #
    num_count += 5 

#키워드 검색 기능의 시작을 알림
print("="*20)
print("\n키워드를 포함한 뉴스 목록 \n")
print("="*20)

#반복문을 통해 사용자가 원하는 키워드의 기사를 계속 검색해보거나 원할 때 종료할 수 있게 구현
while True:
    search=input("\n찾고 싶은 뉴스 키워드를 입력하세요 : ").strip()
    found = False #뉴스 키워드가 저장된 기사 제목들 안에 포함 되어 있는지 확인하기 위한 변수

    for media, articles in news_dict.items():
        for title in articles:
            if search in title:
                found = True #입력한 키워드를 가진 제목이 있으면 True로 업데이트
                print(f"{media} - {title}")

    #입력한 키워드를 가진 기사가 없을 때 실행
    if not found:
        answer = input("\n키워드를 포함한 뉴스가 존재하지 않습니다.\n다시 입력하시겠습니까? [Y/N]")
        if answer =='Y' or answer =='y':
            continue
        else:
            print("키워드로 뉴스 검색 기능을 종료합니다.")
            break

    #입력한 키워드를 포함한 기사들이 있었고 그것을 찾았을 때 다른 키워드로 또 검색할 것인지 질문 후 답에 따라 프로그램 진행
    repeat = input("\n다른 키워드로 검색 해보시겠습니까? [Y/N]")
    
    if repeat =='Y' or repeat =='y':
        continue
    else:
        print("키워드로 뉴스 검색 기능을 종료합니다.")
        break
```
